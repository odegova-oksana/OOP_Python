### Часть 4. Регулярные выражения.

```python
re.findall('(\d+,?\d*)%', s)
```

```python
re.findall('\w+\.\.\.', s)
```

Также можно указывать несколько вариантов повторений {2,3}

### Часть 6. Функция re.compile

есть функция re.findall(pattern, text), где pattern - шаблон поиска, text - текст где надо искать. 

Слоности возникают когда нужно найти шаблоны в нескольких источниках.



```python
text1 = 'Хорваты занервничали и уже к 35-й минуте насобирали пять фолов. Собственно, и шестое, пробивное нарушение правил гостей не заставило себя долго ждать'
text2 = 'При этом Россияне вышли вперед на 18-й минуте благодаря автоголу нападающего соперника Марио Манджукича. На 28-й минуте полузащитник хорватов Иван Перишич сравнял счет. '
text3 = 'увеличили хавбек Поль Погба (59-я минута) и нападающий Килиан Мбаппе (65-я минута). На 69-й минуте Манджукич сократил отставание сборной Хорватии.'

pattern_string = '\d{1,2}\-[йя]'
```

```python
re.findall(pattern_string, text1)
```

```python
re.findall(pattern_string, text2)
```

```python
re.findall(pattern_string, text3)
```

трудоемко так как паттерн преобразуется в объект, что бы это ни значило. Для оптимизации используем re.compile:

```python
pattern = re.compile('\d{1,2}\-[йя]')
pattern.findall(text2)
```

Теперь команда короче и быстрее выполняется компютером. Также этот метод можно использовать с re.sub. Однако проблемы источников не решена.

### ДЗ

Упражнение продолжает практическую работу из последнего видео урока. Для усовершенствования приложения разберитесь, как можно реализовать получение common words с соседних страниц — тех, на которые есть ссылки.
Возможный алгоритм:
* Получить ссылки на соседние страницы.
* Спарсить отдельно соседние страницы.
* Сложить все в один список.

<a href="/wiki/%D0%A1%D1%83%D1%80%D0%BE%D0%BA_(%D0%B7%D0%BD%D0%B0%D1%87%D0%B5%D0%BD%D0%B8%D1%8F)"

```python
import re
#from wiki_requests import get_topic_page

def get_links(topic):
    html_content = get_topic_page(topic)
    links_pattern = re.compile('<a href="/wiki/([A-Z0-9%_()]+)"')
    links = re.findall(links_pattern, html_content)
    return links

def get_topic_text(topic):
    html_content = get_topic_page(topic)
    words = re.findall("[а-яА-Я\-\']{3,}", html_content)
    return words

def get_common_words(topic):
    words_list = get_topic_text(topic)

    links = get_links(topic)
    for link_topic in links:
        words_list.extend(get_topic_text(link_topic))

    rate = {}
    for word in words_list:
        if word in rate:
            rate[word] += 1
        else:
            rate[word] = 1
    rate_list = list(rate.items())
    rate_list.sort(key = lambda x : -x[1])
    return rate_list

def visualize_common_words(topic):
    words = get_common_words(topic)
    for w in words:
        print(w[0])

def main():
    topic = input("Введите тему: ")
    visualize_common_words(topic)
```

### Инкапсуляция

```python
#_step - одно нижнее подчеркивание знак, что это переменную не надо вызывать снаружи класса, но это возможно 
#__step - два нижних подчеркивания - нельзя ее вызвать снаружи

class Bird:
    name = "Unknown"
    distance = 0
    speed = 0
    _step = 1

    def __init__(self, name, speed=100, steps=0):
        self.name = name
        self.speed = speed
        for step in range(steps):
            self.run()

    def call(self):
        print(self.name)

    def run(self, distance=False):
        if not distance:
            distance = self.speed
        self.distance += int(distance*(1 + self._step/10))
        self._step += 1
```

Метод \_\_str\_\_ переопределяет стандартный метод всех классов строкового представления объекта

```python
# Вариант 1 сравнивания объектов
class A:
    def __init__(self, v1):
        self.v1 = v1

    def __str__(self):
        return "hello"

    def compare(self, other):
        return self.v1 > other.v1

a = A(10)
a2 = A(23)

print(a.compare(a2))
```

```python
# Вариант 2 переопределить стандартные методы __gt__, __lt__
class A:
    def __init__(self, v1):
        self.v1 = v1

    def __str__(self):
        return "hello"

    def __gt__(self, other):
        return self.v1 > other.v1

    def __lt__(self, other):
        return self.v1 < other.v1

    def __eq__(self, other):
        return self.v1 == other.v1

a = A(10)
a2 = A(23)

print(a > a2)
print(a < a2)
print(a == a2)
```

```python
class Word:
    def __init__(self, text, part_of_speech):
        self.text = text
        self.part_of_speech = part_of_speech

class Noun(Word):
    __gramma = "Noun"

    def getGramma(self):
        return self.__gramma

class Verb(Word):
    __gramma = "Verb"

    def getGramma(self):
        return self.__gramma

class Sentence:
    def __init__(self, content):
        self.content = content

    def show(self, words_list):
        sentence = ""
        for i in self.content:
            sentence += words_list[i].text + " "
        print(sentence)

    def show_parts(self, words_list):
        parts = set()
        for i in self.content:
            parts.add(words_list[i].getGramma()
                      if isinstance(words_list[i], Verb) or isinstance(words_list[i], Noun)
                      else parts.add(words_list[i].part_of_speech))
        print(parts)

words_list = []
words_list.append(Noun("Привет", "существительное"))
words_list.append(Word("Как", "вопрос"))
words_list.append(Noun("дела", "существительное"))
words_list.append(Word("я", "местоимение"))
words_list.append(Verb("хочу", "глагол"))
words_list.append(Word("у", "предлог"))
words_list.append(Word("с", "предлог"))
words_list.append(Word("тобой", "местоимение"))
words_list.append(Word("тебя", "местоимение"))
words_list.append(Word("в", "предлог"))
words_list.append(Verb("познакомиться", "глагол"))
words_list.append(Word("последнее", "прилагательное"))
words_list.append(Noun("время", "существительное"))

sentence1 = Sentence([0,3,4,6,7,10])
sentence1.show(words_list)
sentence1.show_parts(words_list)

sentence2 = Sentence([1,2,5,8,9,11,12])
sentence2.show(words_list)
sentence2.show_parts(words_list)
```

Параметры героя: сила, скорость перемещения, количество здоровья

Параметры хомяка: сила, запас прочности, скорость перемещения

```python
#player.py
import random

class Player:
    health = 10
    max_health = 10
    position = [2, 0]

    def was_hit(self, damage):
        self.health -= random.choice(range(damage))

    def wait(self):
        if self.health < self.max_health:
            self.health += 1
        print("player's health:", self.health)
```

```python
#hamsters.py
from random import randint

class Hamster:
    health = 3
    position = [0, 0]

    def __init__(self, hid, map):
        self.id = hid
        self.health = 2
        self.position = self.get_clear_position(map)

    def get_clear_position(self, map):
        map_width = len(map.split("\n")[0]) - 1
        map_height = len(map.split("\n")) - 1
        while True:
            coords = [randint(0, map_width), randint(0, map_height)]
            if map.split("\n")[coords[1]][coords[0]] == "*":
                return coords

    def on_shot(self):
        self.health -= 1
        return self.health == 0
```

```python
#game.py
#from player import Player
#from hamsters import Hamster

hamsters_count = 4

class Game:
    map = """****\n****\n****\n****"""
    gameon = True
    happy_message = "You won!"

    def __init__(self):
        self.player = Player()
        self.hamsters = []
        for i in range(hamsters_count):
            self.hamsters.append(Hamster(i+1, self.get_full_map()))

    def add_point(self, position, name, s):
        li = s.split("\n")
        row = li[position[1]]
        row = row[:position[0]] + name + row[position[0] + 1:]
        li[position[1]] = row
        return "\n".join(li)

    def render_map(self):
        s = self.map
        s = self.add_point(self.player.position, "x", s)
        for h in self.hamsters:
            if h.health > 0:
                s = self.add_point(h.position, str(h.hid), s)
        print(s)

    def move_player(self, destination):
        """ destination = w,a,s,d """
        if destination == "s":
            if self.player.position[1] == len(self.map.split("\n")) - 1:
                return False
            self.player.position[1] += 1 # bottom
        if destination == "w":
            if self.player.position[1] == 0:
                return False
            self.player.position[1] -= 1 # top
        if destination == "a":
            if self.player.position[0] == 0:
                return False
            self.player.position[0] -= 1 # left
        if destination == "d":
            if self.player.position[1] == len(self.map.split("\n")[0]) - 1:
                return False
            self.player.position[0] += 1 # right
        self.on_move(destination)

    def get_full_map(self):
        s = self.map
        for h in self.hamsters:
            s = self.add_point(h.position, str(h.hid), s)
        return s

    def get_hamster_on_position(self, coords):
        s = self.get_full_map()
        return s.split("\n")[coords[1]][coords[0]]

    directions = {"w":"s", "s":"w", "a":"d", "d":"a"}

    def on_move(self, direction):
        hamster = self.get_hamster_on_position(self.player.position)
        if not hamster == "*":
            self.player.was_hit(int(hamster))
            if self.player.health <= 0:
                self.gameon = False
                print("game over.....sorry!")
                return False
            print("player's health:", self.player.health)
            killed = self.hamsters[int(hamster) - 1].on_shot()
            if not killed:
                print("wasn't killed")
                self.move_player(self.directions[direction])
            else:
                print(self.hamsters[int(hamster) - 1].id, "was killed")

    def start(self):
        game.render_map()
        while self.gameon:
            if len(self.hamsters) ==0:
                print(self.happy_message)
                return True
            command = input("Insert command: ")
            if command in ["s", "w", "a", "d"]:
                self.move_player(command)
                self.render_map()
            if command == "e":
                self.player.wait()
            if command == "q":
                self.gameon = False


game = Game()
#game.start()
```
